```python
import pandas as pd
import matplotlib.pyplot as plt

file_path = r'C:\Users\GOD\Documents\MelaSales.csv'
df = pd.read_csv(file_path)

print(df.head())

positions = range(len(df))

bar_width = 0.25

plt.bar([p - bar_width for p in positions], df['Week1'], width=bar_width, color='blue', label='Week 1')
plt.bar(positions, df['Week2'], width=bar_width, color='green', label='Week 2')

plt.title("Weekly Sales Data")
plt.xlabel("Day of the Week")
plt.ylabel("Sales")
plt.xticks(positions, df['Day'])  
plt.legend()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

file_path = r'C:\Users\GOD\Documents\MelaSales.csv'
df = pd.read_csv(file_path)

print(df.head())

positions = range(len(df))
bar_width = 0.25

plt.bar([p - bar_width for p in positions], df['Week1'], width=bar_width, color='red', edgecolor='green', linewidth=2, linestyle='--', label='Week 1')
plt.bar(positions, df['Week2'], width=bar_width, color='yellow', edgecolor='green', linewidth=2, linestyle='--', label='Week 2')
plt.bar([p + bar_width for p in positions], df['Week3'], width=bar_width, color='purple', edgecolor='green', linewidth=2, linestyle='--', label='Week 3')

plt.title("Weekly Sales Data")
plt.xlabel("Day of the Week")
plt.ylabel("Sales")
plt.xticks(positions, df['Day'])
plt.legend()

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

file_path = r'C:\Users\GOD\Documents\MelaSales.csv'
df = pd.read_csv(file_path)

plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
plt.hist(df['Week1'], bins=10, color='skyblue', edgecolor='black')
plt.title('Distribution of Week1')
plt.xlabel('Week1 Sales')
plt.ylabel('Frequency')

plt.subplot(1, 3, 2)
plt.hist(df['Week2'], bins=10, color='salmon', edgecolor='black')
plt.title('Distribution of Week2')
plt.xlabel('Week2 Sales')
plt.ylabel('Frequency')

plt.subplot(1, 3, 3)
plt.hist(df['Week3'], bins=10, color='lightgreen', edgecolor='black')
plt.title('Distribution of Week3')
plt.xlabel('Week3 Sales')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()

```

```python
import matplotlib.pyplot as plt

discounts = [15, 40, 20, 25, 50]
sales = [5000, 7000, 9000, 11000, 13000]

plt.figure(figsize=(8, 6))
plt.scatter(discounts, sales, color='blue', marker='o')

plt.title('Relationship Between Discount Offered and Sales Made')
plt.xlabel('Discount Offered (%)')
plt.ylabel('Sales Made ($)')

plt.grid(True)
plt.show()
```
